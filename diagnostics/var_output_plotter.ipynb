```python
import pathlib
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

variable = 'snow_depth'

in_dir = pathlib.Path('../output/3dvar/')
in_file = in_dir / '3dvar.snow_depth_0000.nc'

ds = xr.open_dataset(in_file)
```

```python
nlocs = ds['nlocs']
```

```python
variable + '@ObsValue'
```

```python
obs = ds[variable + '@ObsValue']
ombg = ds[variable + '@ombg']
oman = ds[variable + '@oman']
b = ds[variable + '@hofx']
hofx = ds[variable + '@hofx0']
ana = ds[variable + '@hofx1']
```

```python
# Plot data
fig = plt.figure()
ax = plt.gca()

ax.plot(nlocs, obs, color='red', marker='o', label="obs")
ax.plot(nlocs, hofx, color='blue', label='hofx')
ax.plot(nlocs, ana, color='green', label='ana')

ax.set_title("Var Output Diagnostics", fontsize=15)
ax.set_xlabel('nloc')
ax.set_ylabel(variable)

# add a legend
plt.legend();

# Set figure size
fig.set_size_inches(20,8)
```
